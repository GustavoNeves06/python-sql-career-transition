```python
Contacts = 250
Meetings = 37
Closed_Deals = 9

conversion_rate = (Meetings / Contacts) * 100
close_rate = (Closed_Deals / Meetings) * 100

print(f"Contacts: {Contacts}")
print(f"Meetings: {Meetings}")
print(f"Closed Deals: {Closed_Deals}")
print(f"Conversion Rate: {conversion_rate:.2f}%")
print(f"Close Rate: {close_rate:.2f}%")
```

```python
# Simple sales funnel calculator with basic validation

Contacts = int(input("Enter number of contacts: "))
Meetings = int(input("Enter number of meetings: "))
Closed_Deals = int(input("Enter number of closed deals: "))

if Contacts > 0:
  conversion_rate = (Meetings / Contacts) * 100
else:
  conversion_rate = 0

if Meetings > 0:
  close_rate = (Closed_Deals / Meetings) * 100
else:
  close_rate = 0


print(f"Conversion Rate: {conversion_rate:.2f}%")
print(f"Close Rate: {close_rate:.2f}%")
```
